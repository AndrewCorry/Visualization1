```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('_mpl-gallery')
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
data = pd.read_csv('water.csv', low_memory=False)
data.head()
data.info()
```

```python
data['datetime'].head(273)
```

## Temperature and Height Subplots

```python
data.plot(subplots = True, title='Water Measurements', legend=False);
```

## Water Temperature Simple Plot

```python
data['datetime'] = pd.to_datetime(data['datetime'])
data.plot(title = 'Water Temperature', x='datetime', y='temp', 
                 xlabel = 'Date',
                 ylabel = ' Temperature (C)',
                 grid = True, legend=False );
```

## Water Height Simple Plot

```python
data.plot(
title = 'Water Height',
x='datetime', xlabel='Date',
y='height', ylabel='Height',
grid= True, legend=False  
);
```
